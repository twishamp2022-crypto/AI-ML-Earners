```python
numbers = [1, 2, 3, 4]

for i in range(len(numbers)):
    numbers[i] = numbers[i] + i

print(numbers)
```

```python
marks = [65, 75,85]

total = 0

for m in marks:
    if m >= 70:
        total += m

print(total)
```

```python
nums = [2,4,6]

nums[0] = nums [2]

nums[2] = nums[1]

print(nums)
```

```python
names = ["Amit", "Neha", "Raj"]

longest = names[0]

for name in names:
    if len(name) > len(longest):
        longest = name

print(longest)
```

```python
data = [10, 20, 30]

for i in range(len(data)):
    data[i] = data[i] + 10

print(max(data))

print(min(data))
```

```python
numbers = [5, 10, 15, 20]

total = 0

for i in range(len(numbers)):
    if i % 2 == 0:
        total += numbers[i]

print(total)
```
