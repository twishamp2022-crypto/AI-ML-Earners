# Nested loops

````python
for i in range(5):
    print("i=", i)
````

````python
for i in range(5):
    for j in range(3):
        print("i=", i, "j=", j)
````

````python
for j in range(3):
    print("j=", j)
````

````python
for _ in range(5):
    for __ in range(4):
        print("*", end=" ")
    print()
````

````python
for _ in range(5):
    for __ in range(4):
        print(_+1, end=" ")
    print()
````

````python
for _ in range(5):
    for __ in range(5):
        print(__+1, end=" ")
    print()
````

````python
for _ in range(3):
    for __ in range(1,5):
        print(__*10, end=" ")
    print()
````

````python
for _ in range(1,6):
    for __ in range(_):
        print("*", end=" ")
    print()
````

````python
for _ in range(1,6):
    for __ in range(_):
        print(_, end=" ")
    print()
````

````python
for _ in range(1,6):
    for __ in range(_):
        print(__+1, end=" ")
    print()
````

````python
n=1
for _ in range(1,6):
    for __ in range(_):
        print(n, end=" ")
        n+=1
    print()
````
